```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '')
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', "").str.replace(',', "").astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', "").astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df['rating'].unique()
```

```python
def removepipe(a):
    if a != "|":
        return a
    else:
        return 0 
```

```python
df['rating'] = df['rating'].apply(removepipe)
```

```python
df['rating'].unique()
```

```python
df['rating']
```

```python
df['rating'] = df['rating'].astype(float)
```

```python
df.info()
```

```python
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df.columns
```

```python
df2 = df[['product_name','category', 'discounted_price', 'actual_price', 'discount_percentage',
          'rating', 'rating_count', 'review_title', 'review_content' ]]
```

```python
df2.head()
```

```python
df2.isna().sum()
```

```python
rating_count_mean = df2['rating_count'].mean()
```

```python
df2['rating_count'] = df2['rating_count'].fillna(rating_count_mean)
```

```python
def getcat(a):
    firstcat = a.split('|')
    return firstcat[0]
    
```

```python
df2['category'] = df2['category'].apply(getcat)
```
